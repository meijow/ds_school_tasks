Проверим следующую гипотезу: стоимость недвижимости дороже, если срок сдачи жилья раньше. Проверять будем только ликвидные объекты (процент выкупа больше 10) и самые застраиваемые регионы (чтобы было достаточно данных)

```python
df['objReady100PercDt'] = pd.to_datetime(df['objReady100PercDt'])
```

```python
df['fin_y'] = df['objReady100PercDt'].dt.year % 100
df['fin_m'] = df['objReady100PercDt'].dt.month
```

```python
regs = df.groupby('region')['pdId'].count().sort_values(ascending=False).head(10).index.tolist()
```

```python
dfpr = df[(df['objElemLivingCnt'] > 0) & (df['soldOutPerc'] > 0.1)][['region', 'fin_y', 'objPriceAvg']]
dfpr = dfpr.dropna()
dfpr = dfpr.groupby(['region', 'fin_y']).agg({'objPriceAvg': 'mean'}).dropna().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
for reg in regs:
    ax.plot(dfpr[dfpr['region'] == reg]['fin_y'], dfpr[dfpr['region'] == reg]['objPriceAvg'], label=reg)
ax.legend()
plt.xlim([22, 27])
plt.show()
```

Получается, что гипотеза неверна и нет выгоды покупать жилье на более раннем этапе строительства.
